Memisahkan antara variabel yang mengandung numeric dan object

Grafik gempa di Jawa Barat pada tahun X

```python
tahun_input = int(input("Masukkan tahun: "))
                    
data_gempa_input = data[data['tahun']==tahun_input].reset_index()
data_sorted = data_gempa_input.sort_values(by=['jumlah_gempa_bumi'], ascending=False)[['tahun','nama_kabupaten_kota','jumlah_gempa_bumi','satuan']].reset_index()
data_total_input = data_gempa_input['jumlah_gempa_bumi'].sum()

plt.figure(figsize=(16,4))
plt.bar(data_gempa_input['nama_kabupaten_kota'], data_gempa_input['jumlah_gempa_bumi'])
plt.xticks(range(0, len(data_gempa_input['jumlah_gempa_bumi'])), data_gempa_input['nama_kabupaten_kota'], rotation = 90)
plt.xlabel('Kabupaten/Kota')
plt.ylabel('Jumlah Gempa')
plt.title(f"Grafik Gempa di Jawa Barat tahun {tahun_input}")
plt.show()

print(data_sorted)
print(f'Total Gempa Bumi di Jawa Barat pada tahun {tahun_input} berjumlah {data_total_input} kejadian')
```

Total terjadinya gempa bumi di Jawa Barat pada tahun 2012-2023

```python
data_grup=data.groupby('tahun')['jumlah_gempa_bumi'].sum().reset_index()
data_grup['satuan'] ='KEJADIAN'
data_grup_gempa=data_grup['jumlah_gempa_bumi'].sum()

plt.figure(figsize=(16,4))
plt.bar(data_grup['tahun'], data_grup['jumlah_gempa_bumi'])
plt.plot(data_grup['tahun'], data_grup['jumlah_gempa_bumi'], marker='.', color='red', linewidth=1, label='Trend')
plt.xticks(data_grup['tahun'], rotation = 45)
plt.xlabel('Tahun')
plt.ylabel('Total Gempa')
plt.title("Gempa di Jawa Barat pada tahun 2012 - 2023")

plt.show()
print(data_grup)
print("Total gempa di Jawa Barat pada tahun 2012-2023 adalah",data_grup_gempa,'kejadian')
```
